```python
import os
import requests
from bs4 import BeautifulSoup
import csv
import time
```

```python
file = open("HTMLFOOD.html",encoding='utf-8')
```

```python
data=file.read()
data[0:1000]
```

```python
soup=BeautifulSoup(data,'html.parser')
```

```python
content=soup.prettify()
```

```python
recipe=soup.find_all('div',{'class':"tile-content"})

recipe=recipe[:1000]
print(len(recipe))
```

```python
recipes=[]
print(len(recipes))
```

```python
for dish in recipe:
    title = dish.find('h2',{'class':"title"}).text
    link=dish.find('a', {'href': True})['href']
    recipes.append({'name':title,'link':link})
    
```

```python
print(len(recipes))
```

```python
with open('recipes.csv', 'w', newline='') as csvfile:
  fieldnames = ['name', 'link']
  writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
  writer.writeheader()
  writer.writerows(recipes)
```

```python
def fun(name,keyword):
    for word in keyword:
        if word in name:
            return True
    return False
```
